```python
# Analysis of Property Price Trends and Changes in Purchasing Power in Andalusia Since 2008
# The data on the average salary in Andalusia and the average property prices in both Andalusia and 
# Malaga have been gathered from reliable sources such as INE and notariado.org.


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
salary = pd.read_csv('Avg_salary_Spain_eng.csv', index_col = 'Year').sort_index()
aprt = pd.read_csv('Avg_apt_prices.csv', index_col = 'Year')
malaga_apt = pd.read_csv('vivienda_mlg_anual.csv').set_index('Year')

print("001")
```

```python
# Cleaning the salary dataset

salary.head()
```

```python
# Filtering the DataFrame for measurement type (average) and gender (both genders).

avg_salary = salary[salary['Measurements and percentiles'] == 'Average']
```

```python
avg_salary_both = avg_salary[avg_salary['Gender'] == 'Both genders']
```

```python
# Grouping and filtering the data by location (Andalusia).

avg_salary_both.columns
```

```python
avg_salary_both_and = avg_salary_both.groupby('\nAutonomous communities').get_group('Andalusia')
```

```python
# Converting the total salary column to the format xxx.xx EUR ant to float type.

avg_salary_both_and['Total']=avg_salary_both_and['Total'].str.replace('.','').str.replace(',','.').astype(float)
```

```python
# Calculating the percentage change in salary over the years, using the salary value from 2008 as the reference
reference_salary_2008 = avg_salary_both_and.loc[2008, 'Total']
def percent_change_column(df, value_column, reference):
    df['Change %' ] = (((df[value_column] - reference)/reference)*100).astype(float).round(2)
```

```python
reference_salary_2008 = avg_salary_both_and.loc[2008, 'Total']


percent_change_column(avg_salary_both_and, 'Total', reference_salary_2008)
```

```python
avg_salary_both_and.head()
```

```python
#Cleaning property prices dataset

aprt.head()
```

```python
malaga_apt.head()
```

```python
# Converting the total salary column to the format xxx.xx EUR and to float type.

aprt['Average price(m2)'] = aprt['Average price(m2)'].str.replace(',','').astype(float)
aprt['Average price'] = aprt['Average price'].str.replace(',','').astype(float)
```

```python
# Grouping and filtering the data by location (Andalusia).

aprt_and = aprt.groupby('Location').get_group('Andalusia')
```

```python
# Organizing and filtering the apartments prices dataset based on construction type (new and second-hand properties).


aprt_and_new = aprt_and.groupby('Type').get_group('New')
aprt_and_second = aprt_and.groupby('Type').get_group('Second hand')
```

```python
# Calculating the percentage change in property prices over the years, using the salary value from 2008 as the reference

aprt_and_new_reference_2008 = aprt_and_new.loc[2008, 'Average price(m2)']
aprt_and_second_reference_2008 = aprt_and_second.loc[2008, 'Average price(m2)']
aprt_malaga_reference_2008 = malaga_apt.loc[2008, 'Average price(m2)']

percent_change_column(aprt_and_new, 'Average price(m2)', aprt_and_new_reference_2008)
percent_change_column(aprt_and_second, 'Average price(m2)', aprt_and_second_reference_2008)
percent_change_column(malaga_apt, 'Average price(m2)', aprt_malaga_reference_2008)
```

```python
# Visualizing property price trends over time.


plt.figure(figsize=(10, 6))
plt.plot(aprt_and_second.index, aprt_and_second['Average price(m2)'], label = 'Second hand property in Andalucia', marker='o')
plt.plot(aprt_and_new.index,aprt_and_new['Average price(m2)'], label = 'New property in Andalucia', marker='o')
plt.plot(malaga_apt.index,malaga_apt['Average price(m2)'], label = 'Property in Malaga city', marker='o')


plt.xlabel('Year')
plt.ylabel('Average price(m2)')
plt.title('Changes in property prices per square meter" (m²)')
plt.legend(title='EUR', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(aprt_and_second.index, aprt_and_second['Average price'], label = 'Second hand property in Andalucia', marker='o')
plt.plot(aprt_and_new.index,aprt_and_new['Average price'], label = 'New property in Andalucia', marker='o')
plt.plot(malaga_apt.index,malaga_apt['Average price'], label = 'Property in Malaga city', marker='o')
plt.plot(avg_salary_both_and.index,avg_salary_both_and['Total'], label = 'Average annual salary in Andalucia', marker='o')


plt.xlabel('Year')
plt.ylabel('EUR')
plt.title('Changes in purchased average property prices and annual salary')
plt.legend(title='EUR', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Creating a function to calculate an approx. price of 80 sq.m. apartment

def column_price80(df, price_column_name):
    df['80m2 aprt'] = df[price_column_name] * 80
  
    
```

```python
# Calculating an approx. price for 80 sq.m. apartment in Andalucia and Malaga

column_price80(malaga_apt, 'Average price(m2)')
column_price80(aprt_and_new, 'Average price(m2)')
column_price80(aprt_and_second, 'Average price(m2)')
```

```python
# Creating a function to calculate how many years of average salary in Andalucia are required to be saved to purchase an 80 sq.m. apartment.

#apt_salary_df = pd.merge(malaga_apt, avg_salary_both_and, how='outer', on='Year' )
def years_period(df, property_price_column):
    df['Years period']= (df[property_price_column]/avg_salary_both_and['Total']).round(1)
#apt_salary_df['Years period'] = (apt_salary_df['Average price']/apt_salary_df['Total']).round(1)
#apt_salary_df[['Price apt. 80 sq.m.', 'Total','Years period']]
```

```python
years_period(aprt_and_second,'80m2 aprt')
years_period(aprt_and_new,'80m2 aprt')
years_period(malaga_apt,'80m2 aprt')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(aprt_and_second.index, aprt_and_second['Years period'], label = 'Second hand property in Andalucia', marker='o')
plt.plot(aprt_and_new.index,aprt_and_new['Years period'], label = 'New property in Andalucia', marker='o')
plt.plot(malaga_apt.index,malaga_apt['Years period'], label = 'Property in Malaga city', marker='o')


plt.xlabel('Year')
plt.ylabel('EUR')
plt.title('Property affordability measured by the number of work years required at the average salary in Andalusia')
plt.legend(title='EUR', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Creating a new table combining data from three property price tables.
# The table includes average prices per square meter for apartments in Andalusia, categorized by construction type (new and second-hand apartments),
# and in Malaga (all construction types).
# The "Years" column represents the number of years of savings based on the average annual salary in Andalusia, reflecting property affordability
# over time.

apt_years_df = pd.merge(
    aprt_and_new [['Average price(m2)', 'Years period']],
    aprt_and_second [['Average price(m2)', 'Years period']],
    how='outer', on='Year' )

apt_years_df_all = pd.merge(
    apt_years_df,
    malaga_apt [['Average price(m2)', 'Years period']],
    how='outer', on='Year' )

apt_years_df_all = apt_years_df_all.rename (columns={
    'Average price(m2)_x': 'Avg Price (m2) New aprt',
    'Average price(m2)_y': 'Avg Price (m2) Second hand aprt',
    'Average price(m2)': 'Avg Price (m2) Malaga aprt',
    'Years period_x': 'Years New aprt',
    'Years period_y': 'Years Second hand aprt',
    'Years period': 'Years Malaga aprt'
})
```
